# I. Numpy

### Импортируйте NumPy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
array = np.zeros(10)
array[4] = 1
print(array)
array_2 = np.asarray(array).reshape(-1, 5)
print(array_2)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
array_4 = np.arange(10,50)
print(array_4)
array_5 = np.arange(50)
array_5 = array_4 [::-1]
print(array_5)
array_6=[x for x in array_5 if x % 2 == 0]
print(array_6)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
array_7 = np.arange(9).reshape(3,3)
print(array_7)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
array_8 = np.random.random((4,3,2))
print(array_8)
print(array_8.min())
print(array_8.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
array_9 = np.dot(np.random.random((6,4)), np.random.random((4,3)))
print(array_9)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array_10=np.random.random((7,7))
print(array_10)
print(np.std(array_10))
print(np.mean(array_10))
array_10_norm = (array_10 - array_10. min ())/ (array_10. max () - array_10. min ())
array_10_norm
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
sns_plot = sns.distplot(df['total_bill'])
fig = sns_plot.get_figure()
```

```python
h = df['total_bill'].hist()
fig = h.get_figure()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
x = df['total_bill']    #  x - координаты точек
y = df['tip']   #  y - координаты точек

fig, ax = plt.subplots()

ax.scatter(x, y,
           c = 'blue')    #  цвет точек

ax.set_facecolor('white')     #  цвет области Axes
ax.set_title('Взаимосвязь')     #  заголовок для Axes
ax.set_xlabel('total_bill')
ax.set_ylabel('tip')

fig.set_figwidth(8)     #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```
